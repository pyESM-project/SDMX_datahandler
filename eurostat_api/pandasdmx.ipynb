```python
import pandasdmx as sdmx
import pandas as pd

# Create a client to access Eurostat's NRG_BAL_C dataflow
client = sdmx.Request('ESTAT')

# Fetch the dataflow for the NRG_BAL_C dataset
dataflow_response = client.dataflow('nrg_bal_c')
# dataflow = dataflow_response.dataflow['nrg_bal_c']

# Fetch the structure of the dataset
structure_response = client.get(resource_type='datastructure', resource_id=dataflow_response.structure.id)
structure = structure_response.structure[dataflow_response.structure.id]

# Dictionary to store different components of the structure as DataFrames
structure_dict = {}

# Extract dimensions and their associated codes
for dim in structure.dimensions.aslist():
    dim_name = dim.id
    
    # Get codelist for each dimension (if applicable)
    if dim.local_representation.enumerated is not None:
        codelist_id = dim.local_representation.enumerated.id
        codelist = structure_response.codelist[codelist_id]
        
        # Store codelist as a DataFrame
        dim_df = pd.DataFrame([(code.id, code.name) for code in codelist], columns=[dim_name, 'Description'])
        structure_dict[dim_name] = dim_df

# Output: A dictionary of DataFrames for each dimension or attribute
for key, df in structure_dict.items():
    print(f'\nStructure for {key}:\n', df)


```
